1. Grocery Billing


```python
i1 = int(input('Price of item 1 : '))
q1 = int(input('Quantity  of item 1 : '))
i2 = int(input('Price of item 2 : '))
q2 = int(input('Quantity  of item 2 : '))
total_bill = i1*q1 + i2*q2
print('Total bill : ', total_bill)
```

3. Salary Slip


```python
s = int(input('Salary : '))
hra = 0.2*s
da = 0.1*s
net_salary = s+hra+da
print('Net Salary : ', net_salary)
```

4. Simple Interst

```python
p = int(input('Principle Amount : '))
r = int(input('Rate of Interst : '))
t = int(input('Time Period (in years) : '))
si = (p*r*t)/100
print('Sample Interest : ', si)
```

8. Area of Triangle

```python
m1 = int(input('Enter the length of first side : '))
m2 = int(input('Enter the length of second side : '))
m3 = int(input('Enter the length of third side : '))
s = (m1+m2+m3)/2
import math
area = math.sqrt(s*(s-m1)*(s-m2)*(s-m3))
print('Area of triangle : ', area)
```

9. Paint Cost Estimator

```python
l = int(input('Enter wall lenght (in meters) : '))
h = int(input('Enter wall height ( in meters) : '))
w = 1.5*1.2
wa = l*h
pa = wa-w
cost = pa*25
print('Total area to be painted : ', pa)
print('Total cost for painting : ', cost)
```
